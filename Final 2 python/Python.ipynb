```python
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
from pandas.api.types import is_numeric_dtype
```

```python
pd.read_csv('Data/Womens Clothing E-Commerce Reviews.csv')
```

```python
class DataProcessor:
    def load_data(self):
        data = pd.read_csv('data/Womens Clothing E-Commerce Reviews.csv')
        return data
    def clean_data(self, data):
        # Remove rows with missing values
        data=data.dropna()
        return data
Data_Processor = DataProcessor()
df = Data_Processor.load_data()
cleaned_data = Data_Processor.clean_data(df)
```

```python
class DataAnalyzer:
    def __init__(self, cleaned_data):
        self.cleaned_data = cleaned_data
        
    def get_average(self, column):
            return self.cleaned_data[column].mean()
    def get_mode(self, column):
            return self.cleaned_data[column].mode()
    def get_distribution(self, column):
            return self.cleaned_data[column].std()
    def get_median(self, column):
            return self.cleaned_data[column].median()
Analyzer = DataAnalyzer(cleaned_data)
average = Analyzer.get_average('Age')
print("The average is: ",average)
mode=Analyzer.get_mode('Age')
print("The mode of Ratings is: ",mode)
dist=Analyzer.get_distribution('Rating')
print("The distribution of Rating is: ", dist)
median=Analyzer.get_median("Age")
print("The median of Age is: ",median)
```

```python
class DataVisualizer:
    def __init__(self, cleaned_data):
        self.cleaned_data = cleaned_data
        
    def plot_line_chart(self, X, Y, title, color):
        sns.lineplot(x=X, y=Y, data=self.cleaned_data, color=color)
        plt.title(title)  
        plt.savefig("Pics/" + title + ".png")
        plt.show()
        
    def plot_distribution(self, column, title, palette,kde=False):
        sns.histplot(x=column, data=self.cleaned_data, kde=kde, palette=palette,hue="Rating")
        plt.title(title)  
        plt.savefig("Pics/" + title + ".png")
        plt.show()
        
    def plot_pie(self, column, title):
        Labels = self.cleaned_data[column].unique()
        values = [self.cleaned_data[column].value_counts()[x] for x in Labels]
        plt.pie(x=values, data=self.cleaned_data, labels = Labels)
        plt.title(title)  
        plt.savefig("Pics/"+title+".png")
        plt.show()
        
    def plot_scatter(self, column_x, column_y, title, palette):
        sns.scatterplot(x=column_x, y=column_y, hue="Rating", data=self.cleaned_data, palette=palette)
        plt.title(title)  
        plt.savefig("Pics/" + title + ".png")
        plt.show()
data = Data_Processor.load_data()
cleaned_data = Data_Processor.clean_data(data)
plot = DataVisualizer(cleaned_data)


Plot_line=Visualizer.plot_line_chart("Age","Positive Feedback Count","Plot Line","red")
plot_distribution=Visualizer.plot_distribution("Age","Plot Distribution","Set1",kde=True)
scatter_plot=Visualizer.plot_scatter("Age","Positive Feedback Count","Scatter Plot","Set1")
plot_pie=Visualizer.plot_pie("Rating","Plot pie")
```
